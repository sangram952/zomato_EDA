```python
Ratings=final_df.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
Ratings
```

##observation  
1.whenever rating is between 4.5 to 4.9 ------> Excellent

2.whenever rating is between 4.0 to 4.4 ------> Very Good

3.whenever rating is between 3.5 to 3.9 ------> Good

4.whenever rating is between 2.5 to 3.4 ------> Average

5.whenever rating is between 1.8 to 2.4 ------> Poor




```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)

sns.barplot(x="Aggregate rating",y='Rating Count', data = Ratings)
```

```python
sns.barplot(x="Aggregate rating",y='Rating Count', data = Ratings, hue='Rating color' ,palette= ['white','red','orange','yellow','green','green'])
```

#observation

1.not rated count is very high

2.Maximum number if rating are between2.5 to 3.4

```python
#count plot 
sns.countplot(x= 'Rating color',data = Ratings,palette= ['blue','red','orange','yellow','green','green'])
```
